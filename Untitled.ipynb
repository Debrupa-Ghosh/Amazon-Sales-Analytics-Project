```python
import pandas as pd
df = pd.read_csv("amazon_gif_sales_2023_2024.csv", parse_dates=['order_date'])
df.info()
df.head()
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day'] = df['order_date'].dt.day
df['order_value'] = df['revenue']  # already calculated in synthetic data
df['is_festival_day'] = df['month'].isin([10]).astype(int)  # festival month flag
```

```python
import matplotlib.pyplot as plt

# Monthly revenue trend
monthly = df.groupby(df['order_date'].dt.to_period('M')).agg(total_revenue=('revenue','sum')).reset_index()
monthly['order_date'] = monthly['order_date'].dt.to_timestamp()
plt.figure(figsize=(10,4))
plt.plot(monthly['order_date'], monthly['total_revenue'])
plt.title("Monthly Revenue (2023-2024)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
# Overall revenue by year
rev_by_year = df.groupby('year').agg(total_revenue=('revenue','sum'), total_units=('units','sum'))
print("Revenue By year :",rev_by_year)

# Festival window (example: Oct 1 - Oct 31) by year
fest_df = df[df['month']==10]
fest_by_year = fest_df.groupby('year').agg(total_revenue=('revenue','sum'), total_units=('units','sum'), avg_discount=('discount_pct','mean'))
print("Festival By Year :",fest_by_year)
```

```python
# Define bins (you can adjust thresholds)
bins = [0, 5000, 20000, 50000, 100000]
labels = ['0-5k', '5k-20k', '20k-50k', '50k+']

# Create a new column with revenue ranges
df['revenue_range'] = pd.cut(df['revenue'], bins=bins, labels=labels, include_lowest=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group by category using actual revenue column
disc_revenue = df.groupby('category').agg(
    avg_discount=('discount_pct', 'mean'),   # discount fraction 0-1
    total_revenue=('revenue', 'sum')         # use actual revenue column
).reset_index()

# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(
    disc_revenue['avg_discount']*100,        # convert fraction to %
    disc_revenue['total_revenue'],           # actual revenue in â‚¹
    s=100, color='skyblue', edgecolor='black'
)

plt.title("Avg Discount vs Total Revenue (per Category)")
plt.xlabel("Average Discount (%)")
plt.ylabel("Total Revenue (â‚¹)")

# Annotate each point with category name
for i, txt in enumerate(disc_revenue['category']):
    plt.annotate(
        txt,
        (disc_revenue['avg_discount'][i]*100, disc_revenue['total_revenue'][i]),
        textcoords="offset points", xytext=(0,5), ha='center'
    )

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1ï¸âƒ£ Start fresh with actual revenue
# -------------------------------
# Ensure discount_pct is in fraction (0-1)
df['discount_pct'] = df['discount_pct'].clip(0,1)

# Recalculate revenue
df['revenue'] = df['units'] * df['price'] * (1 - df['discount_pct'])

# Optional: check revenue stats
print("Revenue stats:", df['revenue'].describe())

# -------------------------------
# 2ï¸âƒ£ Group by category
# -------------------------------
disc_revenue = df.groupby('category').agg(
    avg_discount=('discount_pct','mean'),   # fraction 0-1
    total_revenue=('revenue','sum')         # actual revenue in â‚¹
).reset_index()

# Check grouped data
print(disc_revenue)

# -------------------------------
# 3ï¸âƒ£ Scatter plot
# -------------------------------
plt.figure(figsize=(10,6))
plt.scatter(
    disc_revenue['avg_discount']*100,   # convert fraction to %
    disc_revenue['total_revenue'],      # actual revenue
    s=100, color='skyblue', edgecolor='black'
)

plt.title("Avg Discount vs Total Revenue (per Category)", fontsize=14)
plt.xlabel("Average Discount (%)", fontsize=12)
plt.ylabel("Total Revenue (â‚¹)", fontsize=12)

# Annotate each point with category
for i, txt in enumerate(disc_revenue['category']):
    plt.annotate(
        txt,
        (disc_revenue['avg_discount'][i]*100, disc_revenue['total_revenue'][i]),
        textcoords="offset points", xytext=(0,5), ha='center'
    )

plt.tight_layout()
plt.show()
```
